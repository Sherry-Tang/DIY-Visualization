```python
#random color generate
def randomcolor():
    colorArr = ['1','2','3','4','5','6','7','8','9','A','B','C','D','E','F']
    color = ""
    for i in range(6):
        color += colorArr[random.randint(0,14)]
    return "#"+color
```

```python
# Scheme Colors 
def SchemeColors(colormap,n,colorrange=[0.3,0.8]):
    randomlist=np.linspace(colorrange[0],colorrange[1],n)
    colors=[]
    for i in randomlist:
        cmap=plt.get_cmap(colormap)
        color=cmap(i)
        colors.append(color)
    return colors
```

```python
def CuteTwoAxisPlotting(data,index_col,value_col_1,value_col_2,value_col_3=None,xlim=None,title='I am Cute TwoAxis Bar Chart',pad=150,unit1='台',unit2='台',colors=None,text_distance=0.03,bar_width=0,sort=True):
    if colors is None:
        cmaps=['binary', 'gist_yarg', 'gist_gray', 'gray', 'bone', 'pink',
                         'spring', 'summer', 'autumn', 'winter', 'cool', 'Wistia',
                         'hot', 'afmhot', 'gist_heat', 'copper']
        random_list=random.sample(range(0,len(cmaps)),2)
        colors=[]
        for i in random_list:
            color=SchemeColors(cmaps[i],2,colorrange=[0.5,0.8])
            colors.append(color)
        color_1,color_2,color_3=colors[1][0],colors[0][0],colors[1][1]
    else:
        color_1,color_2,color_3=colors[0],colors[1],colors[-1]
    fontname='msyhl.ttc'
    plt.rcParams['font.sans-serif']=['SimHei']
    font_1=FontProperties(fname=font_path+'\\'+fontname,size=38)
    font_2=FontProperties(fname=font_path+'\\'+fontname,size=58)
    font_3=FontProperties(fname=font_path+'\\'+fontname,size=32)
    if sort:
        df=data.sort_values(value_col_1,ascending=True)
    else:
        df=data.copy()
    label=df[index_col]
    value_1=df[value_col_1]
    value_2=df[value_col_2]
    index=range(len(label))
   
    fig,ax=plt.subplots(figsize=[30,25])
    if value_col_3:
        value_3=df[value_col_3]
        font_4=FontProperties(fname=font_path+'\\'+'msyh.ttc',size=32)
        scale_2=max(value_3)/max(value_2)
        scale_1=max(value_3)/max(value_1)
        value_2_sca=value_2*scale_2*0.7
        value_1_sca=value_1*scale_1*0.2
        ax.barh(index,-value_2_sca,bar_width,color=color_2,label=value_col_2)
        ax.barh(index,value_3,bar_width,color=color_3,label=value_col_3)
        ax.barh(index,value_1_sca,bar_width,color=color_1,label=value_col_1)
        if xlim is None:
            plt.xlim(-max(value_2_sca)*1.18,max(value_3)*1.18)  
        else:
            plt.xlim(xlim[0],xlim[1])
        for i ,j,k in zip(value_1,index,value_1_sca):
            plt.text(k/2,j,str(i)+unit1,fontproperties=font_4,ha='center',va='center',color='white')
        for i,j,k in zip(value_2,index,value_2_sca):
            plt.text(-k-text_distance,j,str(round(i*100))+"%",fontproperties=font_3,ha='left',va='center')
        for i ,j,k in zip(value_3,index,value_1_sca):
            plt.text(i+text_distance,j,str(i)+unit1,fontproperties=font_3,ha='right',va='center')
        plt.legend(fontsize=35,ncol=3,bbox_to_anchor=(1.04,1.08))
    else:
        scale_2=max(value_1)/max(value_2)
        value_2_sca=value_2*scale_2*0.7
        ax.barh(index,-value_2_sca,bar_width,color=color_2,label=value_col_2)
        ax.barh(index,value_1,bar_width,color=color_1,label=value_col_1)
        if xlim is None:
            plt.xlim(-max(value_2_sca)*1.18,max(value_1)*1.18)  
        else:
            plt.xlim(xlim[0],xlim[1])
        for i ,j in zip(value_1,index):
            plt.text(i+text_distance,j,str(i)+unit1,fontproperties=font_3,ha='left',va='center')
        for i,j,k in zip(value_2,index,value_2_sca):
            plt.text(-k-text_distance,j,str(i)+unit2,fontproperties=font_3,ha='right',va='center')
        plt.legend(fontsize=30,ncol=2,bbox_to_anchor=(1.04,1.08))    
    ax.yaxis.grid()
    ax.set_axisbelow(True)
    plt.yticks(index,label,fontproperties=font_1)
    plt.xticks(value_1,'')
    ax.tick_params(width=0)
    ax.spines['bottom'].set_linewidth(0)   
    ax.spines['top'].set_linewidth(0)
    ax.spines['right'].set_linewidth(0)
    ax.spines['left'].set_linewidth(0)
    ax.axvline(0,color='black')
    plt.title(title,fontproperties=font_2,pad=pad)
    plt.savefig(saving_path+'//'+title+'.png',transparent=True,bbox_inches='tight')
    plt.show()
```

```python
def TimeMachinePie(data,index_data=None,title='I am a Time Machine',cmaps=None,language='ENG',fontsize=50,radius=1,unit="",legend_loc=[0,0],saving_path=None):
    if cmaps is None:
        cmaps= ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']
    else:
        cmaps=cmaps
    fig,ax=plt.subplots(figsize=[25,20])
    data=data.reset_index().drop('index',1).fillna('Na')
    if language=='ENG':
        plt.rcParams['font.sans-serif']=['Arial']
    else:
        plt.rcParams['font.sans-serif']=['SimHei']
    plt.rcParams.update({'font.size': fontsize*0.48})
    font_2=FontProperties(fname=font_path+'\\'+'msyhl.ttc',size=fontsize*0.8)
    font_3=FontProperties(fname=font_path+'\\'+'msyhl.ttc',size=fontsize)
    total=[]
    explode=[0.01]*len(list(data))
    for i in data:
        data_copy=data.fillna(0).replace('Na',0)
        sum_=data_copy[i].sum()
        total.append(sum_)
    random_list=random.sample(range(0,len(cmaps)),len(list(data)))
    colors=[]
    startangle=[90]
    for i in random_list:
        cmap=plt.get_cmap(cmaps[i])
        color=SchemeColors(cmap,len(list(data)))
        colors.append(color)
    type_=list(data)
    total_labels=['{}:{}'.format(tp,value,unit) for tp,value in zip(type_,total)]
    total_colors=[]
    for i in range(len(colors)):
        color=colors[i][0]
        total_colors.append(color)
    part_1=plt.pie(total,explode=explode,labels=total_labels,radius=radius,shadow=True,colors=total_colors,textprops={'fontproperties':font_2,'color':'black'},labeldistance=1.1,startangle=90)
    labels=[]
    for i,k in zip(range(len(list(data))),range(len(colors[0]))):
        subs=[]
        angle=sum(startangle)
        startangle.append(total[i]/sum(total)*360)
        for j,b in zip(range(len(data)),range(len(colors[1]))):
            if data.iloc[j,i]=='Na':
                continue
            r=radius*(total[i]-sum(subs))/total[i]
            sub=data.iloc[:,i].iloc[j]
            subs.append(sub)
            if index_data is not None:
                index=data.iloc[:,i].index[j]
                label=f"{index_data.loc[index,data.columns[i]]}:{data.iloc[:,i].iloc[j]}"
            else:
                label=f"{data.iloc[:,i].index[j]}:{data.iloc[:,i].iloc[j]}"
            color=[colors[k][b]]
            plt.pie(total[i]/sum(total),labels=" ",radius=r,colors=[colors[k][b]],labeldistance=1.1,startangle=angle)  
            labels.append(label)
    handle,lb = ax.get_legend_handles_labels()
    len_datas=[]
    for i in range(len(list(data))):
        len_data=len([x for x in data.iloc[:,i] if x!='Na'])
        len_datas.append(len_data)
        down=sum(len_datas[0:i])+len(list(data))
        up=sum(len_datas[0:i+1])+len(list(data))
        legend=legend=plt.legend(title=data.columns[i],frameon=False,labels=labels[down-len(list(data)):up-len(list(data))],
                                 handles=handle[down:up],fontsize=fontsize*0.45,bbox_to_anchor=(i*len(data)*0.045+legend_loc[0],legend_loc[1]/radius)) 
        plt.gca().add_artist(legend)
    plt.title(title,fontproperties=font_3)
    if saving_path is None:
        plt.show()
    else:
        plt.savefig(saving_path+'//'+title+'.png',transparent=True,bbox_inches='tight')
        plt.show()  
```

```python
def QueenDiomondPie(data,index_data=None,title='I am Queen Diomond Pie',cmaps=None,language='ENG',fontsize=50,radius=1,unit="",legend_loc=[0,0],saving_path=None):
    if cmaps is None:
        cmaps= ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']
    else:
        cmaps=cmaps
    fig,ax=plt.subplots(figsize=[25,20])
    data=data.reset_index().drop('index',1).fillna('Na')
    if language=='ENG':
        plt.rcParams['font.sans-serif']=['Arial']
    else:
        plt.rcParams['font.sans-serif']=['SimHei']
    plt.rcParams.update({'font.size': fontsize*0.48})
    font_2=FontProperties(fname=font_path+'\\'+'msyhl.ttc',size=fontsize*0.8)
    font_3=FontProperties(fname=font_path+'\\'+'msyhl.ttc',size=fontsize)
    explode=[0.01]*len(list(data))
    total=[]
    for i in data:
        data_copy=data.fillna(0).replace('Na',0)
        sum_=data_copy[i].sum()
        total.append(sum_)
    random_list=random.sample(range(0,len(cmaps)),len(list(data)))
    colors=[]
    startangle=[90]
    for i in random_list:
        cmap=plt.get_cmap(cmaps[i])
        color=SchemeColors(cmap,len(data)+1)
        colors.append(color)
    type_=list(data)
    total_labels=['{}:{}'.format(tp,value,unit) for tp,value in zip(type_,total)]
    total_colors=[]
    for i in range(len(colors)):
        color=colors[i][0]
        total_colors.append(color)
    part_1=plt.pie(total,explode=explode,labels=total_labels,radius=radius,shadow=True,colors=total_colors,textprops={'fontproperties':font_2,'color':'black'},labeldistance=1.1,startangle=90)
    labels=[]
    a=90
    colors=[x[1:] for x in colors]
    index=0
    for i,k in zip(range(len(list(data))),range(len(colors[0]))):
        subs=[]
        for j,b in zip(range(len(data)),range(len(colors[1]))):   
            if data.iloc[j,i]=='Na':
                continue
            index+=1
            data_copy=data.fillna(0).replace('Na',0)
            startangle=a
            a+=data_copy.iloc[j,i]/sum(total)*360
            r=radius*(total[i]-sum(subs))/total[i]
            sub=data.iloc[:,i].iloc[j]
            subs.append(sub)
            if index_data is not None:
                index=data.iloc[:,i].index[j]
                label=f"{index_data.loc[index,data.columns[i]]}:{data.iloc[:,i].iloc[j]}"
            else:
                label=f"{data.iloc[:,i].index[j]}:{data.iloc[:,i].iloc[j]}"
            color=[colors[k][b]]
            plt.pie(data_copy.iloc[j,i]/sum(total),labels=" ",radius=radius*0.8,colors=[colors[k][b]],labeldistance=1.1,startangle=startangle)  
            
            labels.append(label)
    handle,lb = ax.get_legend_handles_labels()
    len_datas=[]
    for i in range(len(list(data))):
        len_data=len([x for x in data.iloc[:,i] if x!='Na'])
        len_datas.append(len_data)
        down=sum(len_datas[0:i])+len(list(data))
        up=sum(len_datas[0:i+1])+len(list(data))
        legend=legend=plt.legend(title=data.columns[i],frameon=False,labels=labels[down-len(list(data)):up-len(list(data))],
                                 handles=handle[down:up],fontsize=fontsize*0.45,bbox_to_anchor=(i*len(data)*0.045+legend_loc[0],legend_loc[1]/radius)) 
        plt.gca().add_artist(legend)
   

    plt.title(title,fontproperties=font_3)
    if saving_path is None:
        plt.show()
    else:
        plt.savefig(saving_path+'//'+title+'.png',transparent=True,bbox_inches='tight')
        plt.show()  
```

```python
def NightingaleRose(data,value_col,index_col,title='I am Nightingale Rose Chart',is_label_show=False,labeldis=0.5,label_color='white',darkness=0.5,cmaps=None,language='ENG',fontsize=50,radius=1,unit="",legend_loc=[0,0],saving_path=None):
    data=data.sort_values(value_col,ascending=False)
    if cmaps is None:
        cmaps= ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']
    else:
        cmaps=cmaps
    fig,ax=plt.subplots(figsize=[25,20])
    data=data.reset_index().drop('index',1).fillna('Na')
    if language=='ENG':
        plt.rcParams['font.sans-serif']=['Arial']
    else:
        plt.rcParams['font.sans-serif']=['SimHei']
    plt.rcParams.update({'font.size': fontsize*0.48})
    font_2=FontProperties(fname=font_path+'\\'+'msyh.ttc',size=fontsize*0.7)
    font_3=FontProperties(fname=font_path+'\\'+'msyhl.ttc',size=fontsize)
    explode=[0.01]*len(list(data))
    total=[]
    index=data[index_col]
    values=data[value_col]

    colors=[]
    rand_list=random.sample(range(0,len(cmaps)),len(data))
    for i in rand_list:
        cmap=plt.get_cmap(cmaps[i])
        color=SchemeColors(cmap,1,colorrange=[darkness,darkness])
        colors.append(color)
    startangle=[90]
    labels=[]
    a=0
    startangle=90
    for i,j in zip(range(len(values)),index):
        a=360*values[i]/sum(values) 
        if is_label_show:
            label=f"{round(100*values[i]/sum(values),1)}"+"%"
        else:
            label=" "
        color=colors[i]
        plt.pie(values[i]/sum(values),labels=[label],radius=radius*values[i]/np.mean(values),
                colors=color,labeldistance=labeldis,startangle=startangle,
               textprops={'fontproperties':font_2,"color":label_color})
        startangle+=a
    plt.title(title,fontproperties=font_3)
    labels=index
    plt.legend(labels=index,fontsize=fontsize*0.45,bbox_to_anchor=(0,0.8),)

    if saving_path is None:
        plt.show()
    else:
        plt.savefig(saving_path+'//'+title+'.png',transparent=True,bbox_inches='tight')
        plt.show()  
```
